```python
def lexical(text, word='Abraham'):
  percentage_occurrence = 100 * text.count(word) / len(text)
  print(f"The word '{word}' appears {percentage_occurrence:.2f}% in the Book of Genesis.")

lexical(text3)
```

```python
def lexical_diversity(text):
  return len(text) / len(set(text))

def percentage(count, total):
  return 100 * count / total

lexical_diversity(text3)
```
